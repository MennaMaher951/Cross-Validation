## Importing the dpendencies

```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
# importing the models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

## Data Collection

```python
data = pd.read_csv("heart_disease_data.csv")
data.head()
```

```python
data.describe()
```

```python
# the destribution of target variable 
data['target'].value_counts()
```

## Splitting the Features & Target

```python
x = data.drop(columns='target',axis=1)
y = data['target']
```

## Train Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
print(x.shape, x_train.shape, x_test.shape)
```

## Comparing the performance of the models 

```python
# list of models
models = [LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
# Model Selection
def compare_models_train_test ():
    for model in models:
        # Training the model
        model.fit(x_train,y_train)
        # Evaluating the model
        test_data_prediction = model.predict(x_test)
        accuracy = accuracy_score(y_test,test_data_prediction)
        print('Accuracy Score of the ',model ,'= ',accuracy)
```

```python
compare_models_train_test ()
```

# Cross Validation

```python
# LogisticRegression
CV_scores_LR = cross_val_score(LogisticRegression(max_iter=1000),x,y,cv=5)
print(CV_scores_LR)
mean_accuracy_LR = sum(CV_scores_LR)/len(CV_scores_LR)
mean_accuracy_LR = mean_accuracy_LR*100
mean_accuracy_LR = round(mean_accuracy_LR,2)
print('Mean Accuracy Score of LogisticRegression: ',mean_accuracy_LR)
print('---------------------------------------------------')

# SVC
CV_scores_SVC = cross_val_score(SVC(kernel='linear'),x,y,cv=5)
print(CV_scores_SVC)
mean_accuracy_SVC = sum(CV_scores_SVC)/len(CV_scores_SVC)
mean_accuracy_SVC = mean_accuracy_SVC*100
mean_accuracy_SVC = round(mean_accuracy_SVC,2)
print('Mean Accuracy Score of SVC: ',mean_accuracy_SVC)
print('---------------------------------------------------')

# KNeighborsClassifier
CV_scores_KNC = cross_val_score(KNeighborsClassifier(),x,y,cv=5)
print(CV_scores_KNC)
mean_accuracy_KNC = sum(CV_scores_KNC)/len(CV_scores_KNC)
mean_accuracy_KNC = mean_accuracy_KNC*100
mean_accuracy_KNC = round(mean_accuracy_KNC,2)
print('Mean Accuracy Score of KNeighborsClassifier: ',mean_accuracy_KNC)
print('---------------------------------------------------')

# RandomForestClassifier
CV_scores_RFC = cross_val_score(RandomForestClassifier(),x,y,cv=5)
print(CV_scores_RFC)
mean_accuracy_RFC = sum(CV_scores_RFC)/len(CV_scores_RFC)
mean_accuracy_RFC = mean_accuracy_RFC*100
mean_accuracy_RFC = round(mean_accuracy_RFC,2)
print('Mean Accuracy Score of RandomForestClassifier: ',mean_accuracy_RFC)
```

```python
# Compare models with cross validation

models = [LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]

def compare_models_CV():
    for model in models:
        CV_Score = cross_val_score(model,x,y,cv=5)
        mean_accuracy = sum(CV_Score)/len(CV_Score)
        mean_accuracy = mean_accuracy*100
        mean_accuracy = round(mean_accuracy,2)
        print('Cross Vlidation Accuracies of: ',model,' = ',CV_Score)
        print('Mean Accuracy Score of: ',model,' = ',mean_accuracy)
        print('----------------------------------------------------------')
```

```python
compare_models_CV()
```
